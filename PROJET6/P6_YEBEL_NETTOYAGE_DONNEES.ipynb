```python
# 8. TRANSFORMONS DES VARIABLES DATE EN TYPE DATETIME, P_CAT EN NUMERIQUE ET AJOUT DES VARIABLES.

# 8-1 CONVERSION DU FORMAT DES COLONNES date & product_categ.

df_data['t_date']         = pd.to_datetime(df_data['t_date'], format='%Y-%m-%d %H:%M:%S')

df_data['p_cat']          = pd.to_numeric(df_data['p_cat'], downcast="integer")

# 8-2 AJOUTONS DE NOUVELLES COLONNES : 6 POUR la date DE SESSION ET 1 POUR L'AGE DES CLIENTS.

df_data['t_year']         = df_data['t_date'].dt.year

df_data['t_month']        = df_data['t_date'].dt.month

df_data['t_day']          = df_data['t_date'].dt.day

# 8-3 DEBUT, FIN DU MOIS: 0, 1, 2, 3.

df_data['t_month_part']   = (df_data['t_date'].dt.day / (df_data['t_date'].dt.days_in_month/4+1)).apply(lambda x: str(x)[0])

df_data['t_hour']         = df_data['t_date'].dt.hour

df_data['t_day_week']     = df_data['t_date'].dt.day_name()

m                         = {'Monday':0, 'Tuesday':1, 'Wednesday':2, 'Thursday':3, 'Friday':4, 'Saturday':5, 'Sunday':6}

df_data['t_day_week_num'] =  df_data['t_day_week'].map(m)

# 8-4 DETERMINONS L'AGE DES ACHETEURS.

df_data['t_buyer_age']    =  df_data['t_year']-df_data['c_birth']
```

```python
# 8-5 PROCEDONS A LA DISCRETISATION DE L'AGE DES ACHETEURS.

# PROCEDONS A LA DISCRETISATION DE L'AGE DES ACHETEURS.

def discretise_age(age):
    tranches = [
        {'bornes': (0,30), 'label': '30-'},
        {'bornes': (31,50), 'label': '31-50'},
        {'bornes': (51,70), 'label': '51-70'},
        {'bornes': (71,120),'label': '71+'}
    ]
    for t in tranches:
        if t['bornes'][0] <= age <= t['bornes'][1]:
            return t['label']
```

```python
# 8-6 APPLIQUONS LA DISCRETISATION SUR LE FICHIER df_data.

df_data['t_buyer_age_range'] = df_data['t_buyer_age'].apply(discretise_age)
```

```python
# 8-7 CALCULONS L'AGE DES CONSOMMATEURS.

df_data['c_age']       = 2023- df_data['c_birth']
```

```python
# 8-8 DISCRETISATION DE L'AGE DES CONSOMMATEURS.

df_data['c_age_range'] = df_data['c_age'].apply(discretise_age)
```

```python
# 8-11 DISCRETISONS LE PRIX.

# 8-11-1 FONCTION DE DISCRETISATION

def discretise_price(price):
    tranches = [
        {'bornes': (0,4),  'label': 'very_cheap 3-'},
        {'bornes': (4,10), 'label': 'cheap 4-9'},
        {'bornes': (10,20), 'label': 'medium 10-19'},
        {'bornes': (20,80), 'label': 'expensive 20-79'},
        {'bornes': (80,500), 'label': 'very_expansive 80+'}
    ]
    for t in tranches:
        if t['bornes'][0] <= price < t['bornes'][1]:
            return t['label']
```

```python
# 8-11-2 APPLICATION DE LA DISCRETISATION DU PRIX.

df_data['p_price_range'] = df_data['p_price'].apply(discretise_price)
```

```python
# 10-CREATION ET EXPLORATION DU DATAFRAME PANIER.

# 10-1 VERIFIONS QU'UNE SESSION CORRESPOND A UN SEUL CLIENT.

temp = transactions.groupby("t_sess_id")["c_id"].nunique()

temp.max()
```

```python
# 10-2 DATAFRAME PANIER

panier = transactions.groupby(["t_sess_id","c_id","c_sex","c_birth"])[["t_date"]].max()

panier.columns = ["t_date_achat"]

# 10-2-1 DATAFRAME TEMP.

temp = transactions.groupby(["t_sess_id","c_id","c_sex","c_birth"])["t_date"].min()

# 10-2-2 COLONNE DUREE_SESSION.

panier["duree_session"] = panier.t_date_achat - temp

# 10-2-3 COLONNE NOMBRE_PRODUITS_PANIER.

panier["nombre_produits_panier"] = transactions.groupby(["t_sess_id","c_id","c_sex","c_birth"])["t_date"].count()

# 10-2-4 COLONNE NOMBRE_PRODUITS_DIFFERENTS.

panier["nombre_produits_differents"] = transactions.groupby(["t_sess_id","c_id","c_sex","c_birth"])["p_id"].nunique()

# 10-2-5 COLONNE MONTANT_PANIER.

panier["montant_panier"] = transactions.groupby(["t_sess_id","c_id","c_sex","c_birth"])["p_price"].sum()
```

```python
# 10-2-6 VERIFIONS L'UTILITE DE LA COLONNE NOMBRE_PRODUITS_DIFFERENTS.

panier[panier.nombre_produits_panier != panier.nombre_produits_differents].count()
```
